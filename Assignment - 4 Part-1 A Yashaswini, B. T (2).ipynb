### 1.Calculate the average growth of states for the duration 2013-14, 2014-15 and 2015-16 by taking the mean of the row '(% Growth over previous year)'. Compare the calculated values and plot them for the states. Make appropriate transformations, if necessary, to plot the data. Report the average growth rates of the various states

```python
selected_years = ['2013-14', '2014-15', '2015-16']
df2 = df.query("Duration in @selected_years")
df2
```

```python
# Filter the DataFrame for the specific condition
df_filtered = df2.loc[df2["Items  Description"] == "(% Growth over previous year)"]

# Select only state columns by excluding "Duration" and "Items  Description"
state_columns = df.columns.difference(["Duration", "Items  Description"])

# Calculate the mean for the filtered DataFrame
average_growth = df_filtered[state_columns].mean()

# Display the result
print("Average percentage growth across each state:")
print(average_growth)
```

```python
import matplotlib.pyplot as plt
```

```python
# Plotting the data
plt.figure(figsize=(18, 15))
average_growth.plot(kind='bar', color='Pink', edgecolor='black')

# Adding labels and title
plt.title("Average Percentage Growth Across States", fontsize=14)
plt.xlabel("States", fontsize=12)
plt.ylabel("Average Growth (%)", fontsize=14)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

### 2.Which states have been growing consistently fast, and which ones have been struggling

```python
# Filter rows for "Percentage Growth" and drop unnecessary columns
growth_df = df2.query("`Items  Description` == '(% Growth over previous year)'").drop(columns=["Items  Description", "Duration"])

# Define a function to check for consistent growth trend
def is_consistently_growing(values):
    # Check if all values are positive and strictly increasing
    positive = (values > 0).all()
    increasing = (values.diff() >= 0).all()  # Using `.diff()` to check increases
    return positive and increasing

# Analyze growth trends for each state
growth_analysis = {
    state: "Consistently Growing" if is_consistently_growing(growth_df[state]) else "Struggling"
    for state in growth_df.columns
}

# Display the results
for state, status in growth_analysis.items():
    print(f"{state}: {status}")
```

### 3.What has been the average growth rate of your home state, and how does it compare to the national average over this duration

```python
# Filter the DataFrame for rows where "Items  Description" is "(% Growth over previous year)"
percentage_growth = df2.query("`Items  Description` == '(% Growth over previous year)'")

# Calculate the average growth rates
growth_stats = percentage_growth[["Karnataka", "All_India GDP"]].mean()

# Extract the growth rates
karnataka_avg_growth = growth_stats["Karnataka"]
india_avg_growth = growth_stats["All_India GDP"]

# Perform the comparison
difference = karnataka_avg_growth - india_avg_growth
ratio = karnataka_avg_growth / india_avg_growth

# Display the results
print(f"Average Growth Rate of Karnataka: {karnataka_avg_growth:.2f}%")
print(f"Average Growth Rate of India: {india_avg_growth:.2f}%")
print(f"Difference in Growth Rate: {difference:.2f}%")
print(f"Ratio of Karnataka to India Growth: {ratio:.2f}")
```

### 4.Identify the top 5 and the bottom 5 states based on total GDP

```python
# Step 1: Filter for year 2015-16 and GSDP
filtered_df = df[(df["Duration"] == '2015-16') & (df["Items  Description"] == "GSDP - CURRENT PRICES (` in Crore)")]
filtered_df.set_index('Items  Description', inplace= True )
```

```python
# Step 2: Drop 'Year' and 'Item Description' columns and transpose
gsdp_data = filtered_df.drop(columns=["Duration"]).transpose()
gsdp = gsdp_data.dropna()
state_gsdp = gsdp.drop('All_India GDP')
```

```python
# Step 3: Sort by GSDP
sorted_gsdp = state_gsdp .sort_values(by= 'GSDP - CURRENT PRICES (` in Crore)', ascending=False)

# Step 4: Select top 5 and bottom 5 states
top_5 = sorted_gsdp.head(5)
bottom_5 = sorted_gsdp.tail(5)

print("Top 5 States by GSDP in 2015-16:")
print(top_5)

print("\nBottom 5 States by GSDP in 2015-16:")
print(bottom_5)
```

```python
# Step 6: Plotting the top 5 and bottom 5 states
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Top 5 States Plot
ax[0].barh(top_5.index, top_5['GSDP - CURRENT PRICES (` in Crore)'], color='Skyblue')
ax[0].set_title('Top 5 States by GSDP (2015-16)')
ax[0].set_xlabel('GSDP')
ax[0].set_ylabel('States')

# Bottom 5 States Plot
ax[1].barh(bottom_5.index, bottom_5['GSDP - CURRENT PRICES (` in Crore)'], color='yellow')
ax[1].set_title('Bottom 5 States by GSDP (2015-16)')
ax[1].set_xlabel('GSDP')
ax[1].set_ylabel('States')

# Show the plots
plt.tight_layout()
plt.show()
```
