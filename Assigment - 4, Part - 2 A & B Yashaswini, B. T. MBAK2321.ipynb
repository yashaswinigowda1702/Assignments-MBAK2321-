## 4.Categorise the states into four groups based on the GDP per capita (C1, C2, C3, C4, where C1 would have the highest per capita GDP and C4, the lowest). The quantile values are (0.20,0.5, 0.85, 1), i.e., the states lying between the 85th and the 100th percentile are in C1; those between the 50th and the 85th percentiles are in C2, and so on.

```python

# Calculate quantiles
quantiles = df1['2014-15'].quantile([0.20, 0.50, 0.85, 1.0])

# Define function to assign categories based on quantile ranges
def categorize_gdp(gdp):
    if gdp >= quantiles[0.85]:
        return 'C1'  # 85th-100th percentile
    elif gdp >= quantiles[0.50]:
        return 'C2'  # 50th-85th percentile
    elif gdp >= quantiles[0.20]:
        return 'C3'  # 20th-50th percentile
    else:
        return 'C4'  # 0-20th percentile

# Apply the function to categorize the states
df1['GDP_Category'] = df1['2014-15'].apply(categorize_gdp)

df1
```

## 5.Find the top sectors (such as agriculture, forestry and fishing, crops, manufacturing etc., not primary, secondary and tertiary) that contribute to approximately 80% of the GSDP of each category

```python
exclude_rows = ['Primary', 'Secondary', 'Tertiary', 'TOTAL GSVA at basic prices',
                'Taxes on Products', 'Subsidies on products', 'Gross State Domestic Product',
                'Population (00)', 'Per Capita GSDP (Rs.)']
df_filtered = df1[~df1['Item'].isin(exclude_rows)]

df_filtered
```

```python
# Clean column names to ensure they are correctly formatted
df_filtered.columns = df_filtered.columns.str.strip()

# Verify column names to check for issues
print("Column names:", df_filtered.columns)
# Group by GDP_category, then by Item to calculate the sum of GSDP (2014-15 column)
gdp_sum = df_filtered.groupby(['GDP_Category', 'Item'], as_index=False)['2014-15'].sum()

# Function to find sectors contributing to 80% of the GSDP for each category
def find_top_sectors_for_80(df_filtered, gdp_column='2014-15', target_percentage=80):
    result = []
    
    # Loop through each GDP category
    for category in df_filtered['GDP_Category'].unique():
        category_data = df_filtered[df_filtered['GDP_Category'] == category]
        
        # Sort sectors by GSDP contribution in descending order
        category_data = category_data.sort_values(by=gdp_column, ascending=False)
        
        # Calculate the cumulative sum of GSDP
        category_data['cumulative_sum'] = category_data[gdp_column].cumsum()
        
        # Calculate total GSDP for the category
        total_gdp = category_data[gdp_column].sum()
        
        # Find sectors contributing to 80% of total GSDP
        threshold = total_gdp * target_percentage / 100
        top_sectors = category_data[category_data['cumulative_sum'] <= threshold]
        
        result.append(top_sectors[['GDP_Category', 'Item', gdp_column]])
    
    # Concatenate the results from each category into a final dataframe
    final_result = pd.concat(result, ignore_index=True)
    return final_result

# Get the top sectors contributing to 80% of the GSDP
top_sectors_df = find_top_sectors_for_80(gdp_sum)

top_sectors_df
```

## 6.Plot the contribution of the sub-sectors as a percentage of the GSDP of each category..

```python
# Group by GDP_category and Item, then calculate the sum of GSDP for each Item within each GDP_category
gdp_sum = df_filtered.groupby(['GDP_Category', 'Item'], as_index=False)['2014-15'].sum()

# Calculate total GSDP for each GDP_category
gdp_sum['Total_GSDP'] = gdp_sum.groupby('GDP_Category')['2014-15'].transform('sum')

# Calculate the percentage contribution of each Item to the total GSDP for that GDP_category
gdp_sum['Percentage_Contribution'] = (gdp_sum['2014-15'] / gdp_sum['Total_GSDP']) * 100

# Plot the contribution as a bar plot
plt.figure(figsize=(20, 10))
for category in gdp_sum['GDP_Category'].unique():
    category_data = gdp_sum[gdp_sum['GDP_Category'] == category]
    plt.bar(category_data['Item'], category_data['Percentage_Contribution'], label=f'Category {category}')

plt.title('Contribution of Sub-Sectors to GSDP by Category')
plt.xlabel('Sub-Sectors (Items)')
plt.ylabel('Percentage Contribution to GSDP')
plt.xticks(rotation=90)
plt.legend(title='GDP Category')
plt.tight_layout()
plt.show()
```

## 8.Analyse if there is any correlation of GDP per capita with dropout rates in education (primary, upper primary and secondary) for the year 2014-2015 for each state. Choose an appropriate plot to conduct this analysis..

```python
edu_df.rename(columns={'Level of Education - State': 'State'}, inplace=True)
```

```python
# Merge the DataFrames on the 'state' column
merged_df = pd.merge(gdp_df, edu_df, on='State')
merged_df
```

```python
# Calculate correlation
correlation = merged_df[['2014-15', 'Primary - 2014-2015','Upper Primary - 2014-2015','Secondary - 2014-2015']].corr()
print("Correlation Matrix:")
print(correlation)
```

```python
# Visualize the correlation with scatter plots
plt.figure(figsize=(15, 5))

for i, col in enumerate(['Primary - 2014-2015','Upper Primary - 2014-2015','Secondary - 2014-2015']):
    plt.subplot(1, 3, i + 1)
    sns.regplot(x='2014-15', y=col, data=merged_df)
    plt.title(f'GDP per Capita vs {col.capitalize()} Dropout')
    plt.xlabel('GDP per Capita')
    plt.ylabel(f'{col.capitalize()} Dropout (%)')

plt.tight_layout()
plt.show()
```

### The correlation matrix indicates that 2014-15 has a strong negative correlation with Upper Primary (-0.651) and Secondary (-0.581), suggesting an inverse relationship. Primary has a moderate positive correlation with Upper Primary (0.541), while Upper Primary and Secondary show a moderate association (0.509). These relationships hint at contrasting trends across different categories.
